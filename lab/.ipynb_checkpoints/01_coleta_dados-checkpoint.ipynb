### **Objetivo:**
- Coletar lista de deputados e seus eventos via API da C√¢mara

```python
import requests
import pandas as pd
from tqdm import tqdm
from datetime import date
import os

# Criar diret√≥rio de sa√≠da - se ainda n√£o tiver/perder diret√≥rio data
os.makedirs("../data/processed", exist_ok=True)
BASE_URL = "https://dadosabertos.camara.leg.br/api/v2"

# Coletar TODOS os deputados com pagina√ß√£o
session = requests.Session()
session.headers.update({
    "Accept": "application/json",
    "User-Agent": "PUCRS-Assiduidade/1.0 (contato: seu_email@exemplo.com)"
})
```

### 1Ô∏è‚É£ Deputados:

```python
deputados_total = []
pagina = 1
while True:
    params_dep = {"itens": 100, "pagina": pagina}
    r = session.get(f"{BASE_URL}/deputados", params=params_dep, timeout=30)
    if r.status_code != 200:
        print(f"‚ö†Ô∏è Falha deputados p√°g {pagina}: {r.status_code}")
        break
    dados = r.json().get("dados", [])
    if not dados:
        break
    deputados_total.extend(dados)
    pagina += 1
    time.sleep(0.15)  # polidez

df_deputados = pd.DataFrame(deputados_total)
df_deputados.to_json("../data/processed/deputados.json", orient="records", indent=2, force_ascii=False)
print(f"‚úÖ Deputados coletados: {len(df_deputados)}")
```

### 2Ô∏è‚É£ Eventos:

- Evento √© uma atividade oficial registrada na C√¢mara dos Deputados e associada a um ou mais parlamentares. Esses eventos servem para medir **frequ√™ncia**, **assiduidade** e **atua√ß√£o**.

```python
eventos_total = []

params_base = {
    "dataInicio": "2020-01-01",
    "dataFim": str(date.today()),
    "itens": 100,
    "ordenarPor": "dataHoraInicio",
    "ordem": "ASC",
}

for dep_id in tqdm(df_deputados["id"], desc="Coletando eventos..."):
    pagina = 1
    while True:
        params = {**params_base, "pagina": pagina}

        # retries simples
        for tentativa in range(3):
            try:
                r = session.get(f"{BASE_URL}/deputados/{dep_id}/eventos", params=params, timeout=30)
                if r.status_code == 200:
                    break
                time.sleep(0.5 + tentativa)  # backoff leve
            except requests.RequestException:
                time.sleep(0.5 + tentativa)
        else:
            print(f"‚ö†Ô∏è Falha ap√≥s 3 tentativas: dep {dep_id} p√°g {pagina}")
            break

        dados = r.json().get("dados", [])
        if not dados:  # acabou a pagina√ß√£o para esse deputado
            break

        # anexa e marca o deputado
        eventos_total.extend([{**e, "id_deputado": int(dep_id)} for e in dados])
        pagina += 1
        time.sleep(0.15)  # polidez

df_eventos = pd.DataFrame(eventos_total)

# Deduplica√ß√£o por seguran√ßa (evento por deputado)
if not df_eventos.empty:
    df_eventos.drop_duplicates(subset=["id", "id_deputado"], inplace=True)
    df_eventos.sort_values(["id_deputado", "id"], inplace=True)

df_eventos.to_json("../data/processed/eventos.json", orient="records", indent=2, force_ascii=False)
print(f"‚úÖ Eventos coletados: {len(df_eventos)}")
```
