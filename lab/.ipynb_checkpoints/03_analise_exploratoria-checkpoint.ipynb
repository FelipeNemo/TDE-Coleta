### **Objetivo:**
- Identificar padr√µes e rela√ß√µes entre presen√ßa, partidos e eventos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar datasets json processados
df_dep = pd.read_json("../data/processed/deputados.json")
df_evt = pd.read_json("../data/processed/eventos.json")

# Converter datas e extrair o ano
df_evt["dataHoraInicio"] = pd.to_datetime(df_evt["dataHoraInicio"], errors="coerce")
df_evt["ano"] = df_evt["dataHoraInicio"].dt.year

# Contar eventos por deputado
freq_eventos = df_evt.groupby("id_deputado").size().reset_index(name="num_eventos")

# Juntar com dados dos deputados (merge)
df_merge = df_dep.merge(freq_eventos, left_on="id", right_on="id_deputado", how="left").fillna(0)

# Estat√≠sticas descritivas 
display(df_merge[["nome", "siglaPartido", "siglaUf", "num_eventos"]].describe())
```

```python
# Histograma: distribui√ß√£o de eventos 
plt.figure(figsize=(9,5))
sns.histplot(df_merge["num_eventos"], bins=20, kde=True)
plt.title("Distribui√ß√£o de Eventos por Deputado")
plt.xlabel("N√∫mero de Eventos")
plt.ylabel("Quantidade de Deputados")
plt.show()

# Top 10 deputados com mais eventos 
top10 = df_merge.sort_values("num_eventos", ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(data=top10, x="num_eventos", y="nome", palette="viridis")
plt.title("Top 10 Deputados com Mais Participa√ß√µes em Eventos")
plt.xlabel("N√∫mero de Eventos")
plt.ylabel("Deputado")
plt.show()
```
