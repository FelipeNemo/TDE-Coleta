### **Objetivo:**
- Descrever campos, tipos, exemplos e significado dos dados coletados.


```python
import pandas as pd

# Ler JSONs processados
df_dep = pd.read_json("../data/processed/deputados.json")
df_evt = pd.read_json("../data/processed/eventos.json")

print("ðŸ“˜ Estrutura â€” Deputados:")
df_dep.info()
print("\nðŸ“— Estrutura â€” Eventos:")
df_evt.info()
```

```python
# Montar tabela descritiva de campos e tipos
descricao = pd.concat([
    pd.DataFrame({
        "Dataset": "deputados",
        "Campo": df_dep.columns,
        "Tipo de Dado": df_dep.dtypes.astype(str)
    }),
    pd.DataFrame({
        "Dataset": "eventos",
        "Campo": df_evt.columns,
        "Tipo de Dado": df_evt.dtypes.astype(str)
    })
])

# Salvar e mostrar amostra
descricao.to_json("../data/processed/descritivo_campos.json", orient="records", indent=2, force_ascii=False)
print("âœ… Descritivo salvo em: data/processed/descritivo_campos.json")
descricao.head(10)
```
